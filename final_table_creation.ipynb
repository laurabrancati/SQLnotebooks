```python
import pandas as pd
import sqlite3
```

```python
states = pd.read_csv("./data/states.csv")
states.dtypes
```

```python
state_ab = []
for state_info in states.values:
    state_ab.append(state_info[1])
print(state_ab)
```

```python
emergency = pd.read_csv("./data/Ballotpedia_StateEmergency.csv")
emergency = emergency.fillna(0)
emergency[["year", "emergencyMonth", "emergencyDay"]] = emergency["StateEmergency"].str.split("-", expand = True)
emergency[["Syear", "schoolClosureMonth", "schoolClosureDay"]] = emergency["FirstSchoolClosure"].str.split("-", expand = True)
emergency = emergency.drop(["StateEmergency", "FirstSchoolClosure", "year","Syear"], axis = 1)
emergency
```

```python
tests = pd.read_csv("./data/CovidDailyTracking_Filtered_Nov16.csv")
tests = tests.fillna(0)
for row in range(len(tests)):
    if tests.loc[row][1] not in state_ab:
        tests = tests.drop(row, axis = 0)
tests[["year", "month", "day"]] = tests["date_clean"].str.split("-", expand = True)
tests = tests.drop(["date_clean", "year"], axis = 1)
tests
```

```python
pop = pd.read_csv("./data/PopDensity.csv")
pop = pop.fillna(0)
pop
```

```python
home = pd.read_csv("./data/StayAtHome_KFF_Clean.csv")
home = home.fillna(0)
home[["Ayear", "monthAnnounced", "dayAnnounced"]] = home["Date Announced"].str.split("-", expand = True)
home[["Eyear", "monthEffective", "dayEffective"]] = home["Effective Date"].str.split("-", expand = True)
home = home.drop(["Date Announced", "Effective Date", "Ayear", "Eyear"], axis = 1)
home
```

## create tables

```python
conn = sqlite3.connect('./data/Covid.db')
curs = conn.cursor()
curs.execute("PRAGMA foreign_keys=ON;")
```

```python
#### state table 
curs.execute("DROP TABLE IF EXISTS tState;")

sql = """CREATE TABLE tState (
            st TEXT PRIMARY KEY CHECK(length(st)==2),
            state TEXT NOT NULL);"""
curs.execute(sql)
```

```python
#emergency table 
curs.execute("DROP TABLE IF EXISTS tEmergency;")

sql = """CREATE TABLE tEmergency (
            st TEXT REFERENCES tState(st),
            emergencyMonth INTEGER NOT NULL,
            emergencyDay INTEGER NOT NULL,
            schoolClosureMonth INTEGER NOT NULL,
            schoolClosureDay INTEGER NOT NULL
            );"""
curs.execute(sql)
```

```python
#testing table
curs.execute("DROP TABLE IF EXISTS tTest;")

sql = """CREATE TABLE tTest (
            st TEXT REFERENCES tState(st),
            month INTEGER NOT NULL,
            day INTEGER NOT NULL,
            positive INTEGER NOT NULL,
            negative INTEGER NOT NULL,
            deaths INTEGER NOT NULL,
            posIncrease INTEGER NOT NULL,
            negIncrease INTEGER NOT NULL,
            deathIncrease INTEGER NOT NULL,
            totalTestResults INTEGER NOT NULL,
            totalTestRessultsIncrease INTEGER NOT NULL);"""
curs.execute(sql)
```

```python
#population table 
curs.execute("DROP TABLE IF EXISTS tPop;")

sql = """CREATE TABLE tPop (
            st TEXT REFERENCES tState(st),
            pop2019 INTEGER NOT NULL,
            popPerSqMi INTEGER NOT NULL);"""
curs.execute(sql)
```

```python
#stay at home order table
curs.execute("DROP TABLE IF EXISTS tHome;")

sql = """CREATE TABLE tHome(
            st TEXT REFERENCES tState(st),
            monthAnnounced INTEGER NOT NULL,
            dayAnnounced INTEGER NOT NULL,
            monthEffective INTEGER NOT NULL,
            dayEffective INTEGER NOT NULL
            );"""
curs.execute(sql)
```

# Load data 

```python
sql = "INSERT INTO tState VALUES (?,?);"
for row in states.values:
    curs.execute(sql, (row[1], row[0]))
```

```python
state_data = pd.read_sql("SELECT * FROM tState;", conn)
state_data
```

```python
#emergency data
emergency_complete = pd.merge(state_data, emergency, left_on = 'state', right_on = "State")
emergency_complete = emergency_complete.drop(["state", "State"], axis = 1)
sql = "INSERT INTO tEmergency VALUES (?,?,?, ?,?);"
for row in emergency_complete.values:
    curs.execute(sql, (row[0], row[1], row[2], row[3], row[4]))
```

```python
emergency_data = pd.read_sql("SELECT * FROM tEmergency;", conn)
emergency_data
```
